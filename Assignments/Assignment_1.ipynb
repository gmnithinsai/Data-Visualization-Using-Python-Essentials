### Question 1  

plot a line plot between a and b:  
a=np.arange(40,50)  
b=np.arange(50,60)  


```python
a = np.arange(40,50)
b = np.arange(50,60)

plt.xlabel('a')
plt.ylabel('b')
plt.title(' Line Plot between a and b') 

plt.plot(a, b,'ro--')
plt.show()

```

### Question 2  

Plot a line plot showing the sales trend in company 1 and 2:  


days = [1,2,3,4,5,6,7] #days of d week  
sales_1 = [160,150,140,145,175,165,180]   #sales of company1  
sales_2 = [70,90,160,150,140,145,175]  #sales of company2  


```python
days = [1,2,3,4,5,6,7]  #days of d week  
sales_1 = [160,150,140,145,175,165,180]   #sales of company1  
sales_2 = [70,90,160,150,140,145,175]  #sales of company2 

plt.figure(figsize = (12,7))


plt.xlabel('Days of Week')
plt.ylabel(' Sales of Company ')
plt.title(' Sales of Companies with respect to each day in a week ')


plt.plot(days, sales_1 , 'go--', label = 'Company - 1')
plt.plot(days, sales_2, 'b*-', label = 'Company - 2')
plt.legend()

```
