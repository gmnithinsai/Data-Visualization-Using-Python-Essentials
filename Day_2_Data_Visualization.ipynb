```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from numpy.random import randn, randint, uniform, sample
```

```python
x = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y1 = [1, 3, 5, 3, 5, 3, 5, 3, 1]
y2 = [2, 4, 6, 4, 6, 4, 6, 4, 2]

plt.plot(x,y1,label='line1')
plt.plot(x, y2, label='line2')
plt.legend()
plt.show()
```

```python
x = [2, 8, 10]
y = [11, 16, 9]

x2 = [3, 9, 11]
y2 = [6, 15, 7]

plt.bar(x, y,label='x,y')
#plt.show()

plt.bar(x2, y2,label='x2, y2')
plt.legend()
plt.show()

```

```python
x = [2, 8, 10]
y = [11, 16, 9]

x2 = [3, 9, 11]
y2 = [6, 15, 7]

plt.barh(x, y,label='x,y')
#plt.show()

plt.barh(x2, y2,label='x2, y2')
plt.legend()
plt.show()

```

```python
x = [1, 2, 3, 4, 5, 6]
y1 = [35, 67, 89, 90, 65, 56]
y2 = [23, 56, 89, 90, 56, 78]

plt.bar(x,y1)
plt.bar(x,y2)

plt.show()
```
