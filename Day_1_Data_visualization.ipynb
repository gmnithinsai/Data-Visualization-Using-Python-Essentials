```python
from numpy.random import randn,randint,uniform,sample
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
plt.plot(x,y,'go--',linewidth=1,markersize=5)
plt.title('Line plot of x and y')
plt.xlabel('X')
plt.ylabel('Y')
```

```python
x=np.arange(0,4*np.pi,0.1)
y = np.sin(x)

plt.title('Sine Waveform')

plt.plot( x, y, color='blue')
plt.show()
```

```python
x = np.arange(0,11)
y = x*x

plt.title('plot of exponential function')
plt.xlabel(' x ')
plt.ylabel(' y = x*X ')

plt.plot(x, y, 'r+--')
plt.show()
```

## Creating sub plots

```python
plt.figure(figsize=(16,6))

plt.subplot(2,2,1)
plt.plot(x, y, 'ro--')

plt.subplot(2,2,2)
plt.plot(x, y, 'b+--')

plt.subplot(2,2,3)
plt.plot(x, y, 'g*--')

plt.subplot(2,2,4)
plt.plot(x, y, 'y>--')
```
