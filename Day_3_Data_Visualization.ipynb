## Categorical Data Plotting

```python
titanic = pd.read_csv(r'E:\Goeduhub_ML_Program_May_20\data\titanic.csv')
```

```python
titanic.head()
```

```python
sns.catplot(x = 'day', y= 'total_bill',data=df)
```

```python
sns.catplot(x = 'Sex', y= 'Fare',data=titanic)
```

```python
sns.catplot(y = 'Fare', x= 'Pclass',data=titanic)
```

```python
sns.catplot(y = 'Age', x= 'Embarked',data=titanic)
```

```python
sns.catplot(y = 'Age', x= 'Embarked',data=titanic,kind='swarm')
```

```python
sns.catplot(y = 'Age', x= 'Embarked',data=titanic, kind = 'strip')
```

```python
sns.catplot(x='smoker', y='tip', data=df, order = ['No','Yes'])
```

```python
sns.catplot(x = 'day', y = 'tip', kind = 'box', data = df, hue = 'sex')
```

```python
sns.catplot(x = 'day', y = 'tip', kind = 'boxen', data = df.sort_values('day'), hue = 'sex')
```

```python
sns.catplot(x = 'day', y = 'tip', kind = 'violin', data = df, hue = 'sex')
```
